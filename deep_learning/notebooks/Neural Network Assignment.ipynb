```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Sample Data
x1, x2 = 3, 2
y_actual = 17

# Initialize weights randomly between 1 and 10
np.random.seed(42)
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)

learning_rate = 0.01
epochs = 20
errors = []
preds = []

for epoch in range(epochs):
    # Feed Forward
    y_pred = x1 * w1 + x2 * w2
    
    # Error Calculation
    error = (y_actual - y_pred)**2
    errors.append(error)
    preds.append(y_pred)
    
    # Back Propagation
    # Gradient of error w.r.t w1: 2 * (y_pred - y_actual) * x1
    # Gradient of error w.r.t w2: 2 * (y_pred - y_actual) * x2
    grad_w1 = 2 * (y_pred - y_actual) * x1
    grad_w2 = 2 * (y_pred - y_actual) * x2
    
    # Weight update
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

# Results for Markdown Tables
results = {'epoch': list(range(epochs)), 'error': errors, 'ypred': preds}
import pandas as pd
results_df = pd.DataFrame(results)
results_df.to_csv('results.csv', index=False)

# Show last values for reporting
(w1, w2, y_pred, error)

# Prepare error and prediction lists for charting
errors, preds

# Don't show charts here, will do in next step with chart tool.
```

```python
# -------------------------------------------------------------
# Neural Network for Simple Linear Regression (Manual Backprop)
# Objective: Predict y = 3x1 + 4x2 using a simple neural model
# -------------------------------------------------------------

import random
import math
import matplotlib.pyplot as plt

# -------------------------------------------------------------
# 1. Activation Functions (for demonstration)
# -------------------------------------------------------------
def sigmoid(x):
    """Sigmoid activation function"""
    return 1 / (1 + math.exp(-x))

def tanh(x):
    """Tanh activation function"""
    return math.tanh(x)

def relu(x):
    """ReLU activation function"""
    return x if x > 0 else 0

# -------------------------------------------------------------
# 2. Sample Data
# -------------------------------------------------------------
x1 = 3.0
x2 = 2.0
y_actual = 17.0   # Expected: 3*3 + 4*2 = 9 + 8 = 17

# -------------------------------------------------------------
# 3. Initialize Random Weights
# -------------------------------------------------------------
random.seed(42)
w1 = random.uniform(1, 10)
w2 = random.uniform(1, 10)

# -------------------------------------------------------------
# 4. Hyperparameters
# -------------------------------------------------------------
learning_rate = 0.01
epochs = 20

# -------------------------------------------------------------
# 5. Containers for Visualization
# -------------------------------------------------------------
errors = []
ypreds = []

# -------------------------------------------------------------
# 6. Training Loop (Feedforward + Backpropagation)
# -------------------------------------------------------------
for epoch in range(1, epochs + 1):
    
    # Feed Forward Step
    y_pred = x1 * w1 + x2 * w2
    
    # Compute Error (Squared Error)
    error = (y_actual - y_pred) ** 2
    
    # Save for visualization
    errors.append(error)
    ypreds.append(y_pred)
    
    # Compute Gradients (Backpropagation)
    grad_w1 = -2 * (y_actual - y_pred) * x1
    grad_w2 = -2 * (y_actual - y_pred) * x2
    
    # Update Weights
    w1 = w1 - learning_rate * grad_w1
    w2 = w2 - learning_rate * grad_w2
    
    # Print progress
    print(f"Epoch {epoch:02d}: y_pred={y_pred:.6f}, error={error:.6f}, w1={w1:.6f}, w2={w2:.6f}")

# -------------------------------------------------------------
# 7. Final Results
# -------------------------------------------------------------
final_y_pred = x1 * w1 + x2 * w2
final_error = (y_actual - final_y_pred) ** 2

print("\nFinal Results:")
print(f"  Final Weights -> w1: {w1:.6f}, w2: {w2:.6f}")
print(f"  Final Prediction (y_pred): {final_y_pred:.6f}")
print(f"  Actual y: {y_actual}")
print(f"  Final Squared Error: {final_error:.6f}")

# -------------------------------------------------------------
# 8. Visualization
# -------------------------------------------------------------

# Plot 1: Error across epochs
plt.figure(figsize=(8, 4))
plt.plot(range(1, epochs + 1), errors, marker='o', color='red')
plt.title("Error vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("Squared Error")
plt.grid(True)
plt.show()

# Plot 2: y_pred progression
plt.figure(figsize=(8, 4))
plt.plot(range(1, epochs + 1), ypreds, marker='o', color='blue', label='Predicted y')
plt.axhline(y=y_actual, color='green', linestyle='--', label='Actual y')
plt.title("Predicted y vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("y_pred")
plt.legend()
plt.grid(True)
plt.show()
```
